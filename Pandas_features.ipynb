### Бинаризация

```python
diam.assign(price_cat=pd.cut(diam['price'], bins=3, labels=['cheap', 'medium', 'pricy']))  #бинаризация в столбце
```

### Создание столбцов

```python
new5 = diam.assign(ln_depth = np.log(diam['depth'])) #создание столбца
new5
```

```python
diam['cut2'] = diam['cut'].str.contains('Good') #бинарная переменная проверяющая наличие значения в столбце
diam
```

### Создание аггрегированных таблиц

```python
#создание агрегированной таблицы для каждого значения параметра по всем интересующим нас параметрам
diam2 = (diam.groupby('cut')
    .agg(av_x=('x', np.mean),
        min_x=('x', np.min),
        max_weight=('carat', np.max))
    .reset_index())
diam2
```

```python
diam.groupby('color')['depth'].agg(['mean','median'])
```

```python
diam.groupby('color')['x'].mean().sort_values()
```

```python
#можно использовать матрицу зависимостей, чтобы сравнивать некоторую статистику параметра для мультиклассовых переменных
diam.pivot_table(columns='color', index='cut', values='x', aggfunc='mean')
```
